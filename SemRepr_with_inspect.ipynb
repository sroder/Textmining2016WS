```python
import sympy
import inspect
from inspect import *
```

```python
#Create a list with all functions and a list with all public functions

everything = inspect.getmembers(sympy)
all_functions = []
all_public_functions = []

for e in everything:
    arg_string = 'sympy.' + e[0]
    if inspect.isfunction(eval(arg_string)):
        all_functions = all_functions + [e]
        if inspect.getsourcelines(eval(arg_string))[0][0].startswith('@public'):
            all_public_functions = all_public_functions + [e]
```

```python
#Get all important data about a function:

def get_information(arg):
    source = inspect.getsourcelines(arg)[0]
    
    #Get the descriptiontext of the functions:
    
    description = []
    for line in inspect.getsourcelines(arg)[0]:
        if ('Examples' not in line):
            description = description + [line]
        else:
            break
            
    #Clean the description:
    description_modified = [] #description[2:]
    for line in description[2:]:
        if 'Parameters' not in line:
            description_modified = description_modified + [line]
        else:
            break
            
    description_string = ''
    for line in description_modified:
        description_string = description_string + line  
        
    description_string = description_string.replace('    ', ' ')
    description_string = description_string.replace('\n', '')
    description_string = description_string.replace('"""', '')
    description_string = description_string.lstrip()
    

    #Find the parameters:
    
    parameters_as_string = description[1].partition('(')[-1].rpartition(')')[0]
    parameter_list = parameters_as_string.split(', ')

    #Get the descriptions for every parameter and the descipton for every function:
    #ToDo: second line of the description
    
    parameter_description = ()
    parameter_description_list = []
    function_description = []
    
    if '    Parameters\n' in description:
        for parameter in parameter_list:
        
            if '=' in parameter:
                parameter = parameter.split('=')[0]
            if parameter.startswith('*'):
                parameter = parameter.replace('*', '')
            if parameter.startswith('**'):
                parameter = parameter.replace('**', '')
    
            for line in description:
                if line.startswith('    ' + parameter + ' : '):
                    parameter_description_list = parameter_description_list + [(parameter, line.split(' : ')[1].replace('\n', ''))]
    else:
        for parameter in parameter_list:
            if '=' in parameter:
                parameter = parameter.split('=')[0]
            if parameter.startswith('*'):
                parameter = parameter.replace('*', '')
            if parameter.startswith('**'):
                parameter = parameter.replace('**', '')
            
            parameter_description_list = parameter_description_list + [(parameter, None)]
        
     
    #Organize parameters in dictionary:
    
    dict_for_params = {}
    param_dict_list = []
    
    for element in parameter_description_list:
        dict_for_params = {
            'name' : element[0],
            'desc' : element[1]
        }
        param_dict_list = param_dict_list + [dict_for_params]
        
    #Get the name of the function:
    name = description[1].split(' ')[1].split('(')[0]
    
    #Find the data the function returns:
    return_text = []
    for line in source:
        if line.startswith('    Returns'):
            start_index = source.index(line)
            considered_line = line
            while considered_line != '\n':
                return_text = return_text + [considered_line]
                start_index = start_index + 1
                considered_line = source[start_index]
    
    #Get an example:
    if '    Examples\n' in source:
        index = source.index('    Examples\n')
        example = source[index:]
        
        #Clean the example:
        example = example[2:]
        example_string = ''
        for line in example:
            example_string = example_string + line
        
        example_string = example_string.replace('    ', ' ')
        example_string = example_string.replace('\n', '')
        example_string = example_string.lstrip()
    else:
        example = None
        example_string = 'no example available'
    
    #example_string = inspect.cleandoc(example_string)
    
    #Organize everything in a dictionary:
    
    dict_for_func = {
        'name' : name,
        'desc' : description_string,
        'params' : param_dict_list,
        'data' : return_text,
        'example': example_string
    }
    
    return dict_for_func
    
```
